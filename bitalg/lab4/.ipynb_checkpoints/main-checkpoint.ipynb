# Generowanie losowych odcinków na płaszczyźnie

Uzupełnij funkcję ```generate_uniform_sections```

````python
import random
def generate_uniform_sections(max_x, max_y, n):   
    sections = []
    
    for _ in range(n):
        x1 = random.uniform(0, max_x)
        y1 = random.uniform(0, max_y)
        
        x2 = random.uniform(0, max_x)
        y2 = random.uniform(0, max_y)
        
        while x1 == x2:
            x2 = random.uniform(0, max_x)
        
        section = ((min(x1, x2), y1), (max(x1, x2), y2))
        
        sections.append(section)
    
    return sections
````

Uzupełnij funkcję ```is_intersection```

````python
def det(p, q, r):
    return (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])
    
def orientation(p, q, r):
    val = det(p, q, r)
    if val == 0:
        return 0
    return 1 if val > 0 else 2
````

````python
def on_segment(p, q, r):
    return (q[0] <= max(p[0], r[0]) and q[0] >= min(p[0], r[0]) and
            q[1] <= max(p[1], r[1]) and q[1] >= min(p[1], r[1]))

def do_intersect(p1, q1, p2, q2):
    o1 = orientation(p1, q1, p2)
    o2 = orientation(p1, q1, q2)
    o3 = orientation(p2, q2, p1)
    o4 = orientation(p2, q2, q1)

    if o1 != o2 and o3 != o4:
        return True  # przecinają się

    if o1 == 0 and on_segment(p1, p2, q1):
        return True  # p2 leży na p1q1
    if o2 == 0 and on_segment(p1, q2, q1):
        return True  # q2 leży na p1q1
    if o3 == 0 and on_segment(p2, p1, q2):
        return True  # p1 leży na p2q2
    if o4 == 0 and on_segment(p2, q1, q2):
        return True  # q1 leży na p2q2

    return False  # nie przecinają się
    
def is_intersection(sections):
    events = []
    for section in sections:
        events.append((section[0], 'start', section))
        events.append((section[1], 'end', section))

    events.sort()

    active_sections = set()
    intersections = []

    for event in events:
        point, event_type, section = event
        if event_type == 'start':
            for active_section in active_sections:
                if do_intersect(active_section[0], active_section[1], section[0], section[1]):
                    return True
            active_sections.add(section)
        elif event_type == 'end':
            active_sections.remove(section)

    return False
    
    """
    Funkcja sprawdza czy jakakolwiek para podanych odcinków się przecina 
    :param sections: tablica odcinków w postaci krotek krotek współrzędnych punktów końcowych odcinków
    :param eps: epsilon będący dokładnością, z jaką porównujemy liczby zmiennoprzecinkowe
    :return: True / False
    """
````

Uzupełnij funkcję ```find_intersections```

````python
def do_intersect(p1, q1, p2, q2, id1, id2):
    o1 = orientation(p1, q1, p2)
    o2 = orientation(p1, q1, q2)
    o3 = orientation(p2, q2, p1)
    o4 = orientation(p2, q2, q1)

    if o1 != o2 and o3 != o4:
        return True, (id1, id2)
    
    if o1 == 0 and on_segment(p1, p2, q1):
        return True, (id1, id2)

    if o2 == 0 and on_segment(p1, q2, q1):
        return True, (id1, id2)

    if o3 == 0 and on_segment(p2, p1, q2):
        return True, (id1, id2)

    if o4 == 0 and on_segment(p2, q1, q2):
        return True, (id1, id2)

    return False, None

def calculate_slope_and_intercept(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    slope = (y2 - y1) / (x2 - x1)
    intercept = y1 - slope * x1
    return slope, intercept

def find_intersection_point(line1, line2):
    m1, b1 = calculate_slope_and_intercept(line1[0], line1[1])
    m2, b2 = calculate_slope_and_intercept(line2[0], line2[1])

    if m1 == m2:
        return None

    x_intersection = (b2 - b1) / (m1 - m2)
    
    y_intersection = m1 * x_intersection + b1

    return x_intersection, y_intersection

def find_intersections(sections):
    events = []
    for i, section in enumerate(sections):
        events.append((section[0], 'start', i))
        events.append((section[1], 'end', i))

    events.sort()

    active_segments = set()
    intersections = []

    for event in events:
        point, event_type, segment_id = event
        segment = sections[segment_id]

        if event_type == 'start':
            for active_segment_id in active_segments:
                active_segment = sections[active_segment_id]
                intersect, intersected_segments = do_intersect(
                    active_segment[0], active_segment[1], segment[0], segment[1], active_segment_id, segment_id
                )
                if intersect:
                    x, y = find_intersection_point((active_segment[0], active_segment[1]), (segment[0], segment[1]))
                    point = (x, y)
                    intersections.append((point, *intersected_segments))
            
            active_segments.add(segment_id)
        elif event_type == 'end':
            active_segments.remove(segment_id)

    return intersections
    """
    Funkcja znajduje wszystkie przecięcia zadanych odcinków
    :param sections: tablica odcinków w postaci krotek ktorek współrzędnych punktów końcowych odcinków
    :return: tablica punktów przecięć w postaci trzyelementowych krotek w których pierwszym elementem są współrzędne danego punktu, a drugim i trzecim indeksy prostych z listy wejściowej, które się przecinają w tym punkcie współrzędnych np.:
    [((x1, y1), id1,id2),
    ((x2, y2), id3, id4),
    ...]
    """
    raise Exception("Unimplemented")
````

````python
# Zaimplementowany algorytm o lepszej złożoności (z wykładu), niestety popełniłem gdzieś błąd w związku z kluczami SortedDict,
# przez co funkcja del nie działa poprawnie, oraz nie jestem pewny czy __hash__ oraz __eq__ są poprawne
# Jednak w na podstawie debugowania w Pycharmie (przynajmniej próby) mogę stwierdzić, że wyniki samego algorytmu są prawidłowe, więc to tylko problem
# w samej implementacji usuwania elementu ze słowanika


from sortedcontainers import SortedDict
from bisect import insort_left
class Event:
    def __init__(self, x, segment, event_type):
        self.x = x
        self.segment = segment
        self.event_type = event_type

    def __hash__(self):
        # Use the hash of a tuple of the attributes as the hash value
        return hash((self.x, self.segment, self.event_type))

    def __eq__(self, other):
        # Check if two events are equal based on their attributes
        if not isinstance(other, Event):
            return False
        return (
            self.x == other.x
            and self.segment == other.segment
            and self.event_type == other.event_type
        )


def det(p, q, r):
    return (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])


def orientation(p, q, r):
    val = det(p, q, r)
    if val == 0:
        return 0
    return 1 if val > 0 else 2


def on_segment(p, q, r):
    return (max(p[0], r[0]) >= q[0] >= min(p[0], r[0]) and
            max(p[1], r[1]) >= q[1] >= min(p[1], r[1]))


def do_intersect(p1, q1, p2, q2):
    o1 = orientation(p1, q1, p2)
    o2 = orientation(p1, q1, q2)
    o3 = orientation(p2, q2, p1)
    o4 = orientation(p2, q2, q1)

    if o1 != o2 and o3 != o4:
        return True  # przecinają się

    if o1 == 0 and on_segment(p1, p2, q1):
        return True  # p2 leży na p1q1
    if o2 == 0 and on_segment(p1, q2, q1):
        return True  # q2 leży na p1q1
    if o3 == 0 and on_segment(p2, p1, q2):
        return True  # p1 leży na p2q2
    if o4 == 0 and on_segment(p2, q1, q2):
        return True  # q1 leży na p2q2

    return False  # nie przecinają się


def calculate_slope_and_intercept(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    slope = (y2 - y1) / (x2 - x1)
    intercept = y1 - slope * x1
    return slope, intercept


def find_intersection_point(line1, line2):
    m1, b1 = calculate_slope_and_intercept(line1[0], line1[1])
    m2, b2 = calculate_slope_and_intercept(line2[0], line2[1])

    if m1 == m2:
        return None

    x_intersection = (b2 - b1) / (m1 - m2)

    y_intersection = m1 * x_intersection + b1

    return x_intersection, y_intersection

def intersect(seg1, seg2):
    if seg1 == seg2: return False
    if seg1 and seg2:
        p1, q1 = seg1
        p2, q2 = seg2
        return do_intersect(p1, q1, p2, q2)
    return False


def find_intersections2(segments):
    event_queue = SortedDict()
    sweep_line = SortedDict()  # SortedDict to maintain the active segments sorted by their y-coordinate

    for segment in segments:
        event_queue[segment[0][0]] = Event(segment[0][0], segment, 'left')
        event_queue[segment[1][0]] = Event(segment[1][0], segment, 'right')

    output_list = []

    while event_queue:
        event_idx, event = event_queue.popitem(index=0)

        if event.event_type == 'left':
            segE = event.segment
            sweep_line[segE[1][1]] = segE

            # Find adjacent segments in the sweep line
            t, segA = sweep_line.peekitem(sweep_line.bisect_left(segE[1][1]))
            t, segB = sweep_line.peekitem(sweep_line.bisect_right(segE[1][1])-1)

            # Check for intersections and update event queue
            if segA and segB and intersect(segB, segA):
                point = find_intersection_point(segB, segA)
                del event_queue[point]
            if intersect(segE, segA):
                point = find_intersection_point(segE, segA)
                event_queue[event_idx] = Event(point[0], point, (segE, segA))
            if intersect(segE, segB):
                point = find_intersection_point(segE, segB)
                event_queue[event_idx] = Event(point[0], point, (segE, segB))

        elif event.event_type == 'right':
            segE = event.segment
            del sweep_line[segE[1][1]]

            # Check for intersections and update event queue
            t, segA = sweep_line.peekitem(sweep_line.bisect_left(segE[1][1]) - 1)
            t, segB = sweep_line.peekitem(sweep_line.bisect_right(segE[1][1]))
            if segA and segB and intersect(segA, segB):
                point = find_intersection_point(segA, segB)
                event_queue[point[0]] = Event(point[0], point, (segA, segB))

        else:  # Intersection event
            output_list.append(event.segment)
            segE1, segE2 = event.event_type
            del sweep_line[segE1[1][1]]
            del sweep_line[segE2[1][1]]
            sweep_line[segE2[1][1]] = segE2
            sweep_line[segE1[1][1]] = segE1

            # Check for intersections and update event queue
            t, segA = sweep_line.peekitem(sweep_line.bisect_left(segE2[1][1])-1)
            t, segB = sweep_line.peekitem(sweep_line.bisect_right(segE1[1][1]))
            if segA and intersect(segE1, segA):
                point = find_intersection_point(segE1, segA)
                del event_queue[point[0]]
            if segB and intersect(segE2, segB):
                point = find_intersection_point(segE2, segB)
                del event_queue[point[0]]
            if intersect(segE2, segA):
                point = find_intersection_point(segE2, segA)
                event_queue[point[0]] = Event(point[0], point, (segE2, segA))
            if intersect(segE1, segB):
                point = find_intersection_point(segE1, segB)
                event_queue[point[0]] = Event(point[0], point, (segE1, segB))

    return set(output_list)
````
